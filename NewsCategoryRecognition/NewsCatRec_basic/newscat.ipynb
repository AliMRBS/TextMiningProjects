```python
import pandas as pd
```

```python
data = pd.read_csv("../per.csv")
data.head()
```

deleting stop words

```python
with open('stopwords.txt', encoding='utf-8') as stopwords_file:
    stopwords = stopwords_file.read().splitlines()
stopwords
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
nltk_stopwords = nltk.corpus.stopwords.words('english')
stopwords.extend(nltk_stopwords)
len(stopwords)
```

```python
from hazm import Stemmer
stemmer = Stemmer()
```

```python
stemmer.stem('کتاب ها')
```

```python
from hazm import Lemmatizer
lem = Lemmatizer()
```

```python
lem.lemmatize("رفتم")
```

```python
from hazm import word_tokenize
```

```python
dataset = pd.DataFrame(columns=('title_body', 'category'))
for index, row in data.iterrows():
    title_body = row['Title'] + ' ' + row['Body']
    title_body_tokenized = word_tokenize(title_body)
    title_body_tokenized_filtered = [w for w in title_body_tokenized if not w in stopwords]
    title_body_tokenized_filtered_stemmed = [stemmer.stem(w) for w in title_body_tokenized_filtered]
    title_body_tokenized_filtered_lemed = [lem.lemmatize(w).replace('#', ' ') for w in title_body_tokenized_filtered]
    dataset.loc[index] = {
        'title_body': ' '.join(title_body_tokenized_filtered_lemed) + ' ' + ' '.join(title_body_tokenized_filtered_stemmed),
        'category': row['Category2'].replace('\n', '')
    }
```

```python
dataset.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
vectorizer.fit(dataset['title_body'])
```

```python
X = vectorizer.transform(dataset['title_body'])
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(dataset['category'])
```

```python
y
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y)
```

ساخت مدل  با استفاده از svm

```python
from sklearn import svm 
```

```python
svmc = svm.SVC()
svmc.fit(x_train, y_train)
```

```python
svmc.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_pred = svmc.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

ساخت مدل با استفاده از XGBoost

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
model2 = GradientBoostingClassifier(
    n_estimators=10,
    learning_rate=0.1,
    max_depth=2,
    random_state=42,
    max_features="sqrt"
)
```

```python
model2.fit(x_train, y_train)
```

```python
y_pred2 = model2.predict(x_test)
```

```python
accuracy_score(y_test, y_pred2)
```

```python
print(classification_report(y_test, y_pred2))
```

```python
print(confusion_matrix(y_test, y_pred2))
```
